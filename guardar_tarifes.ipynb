# Tarifes AMB: Desplaçament unipersonal

Avui s'han fet públiques les tarifes del transport públic a l'Àrea Metropolitana de Barcelona (AMB) pel 2020. Aquestes tarifes aposten per títols unipersonals i per un us habitual i regular del **transport públic** (a diferència d'apostar per mètodes de transport sostenibles).

```python
import numpy as np
import xarray as xr
xr.set_options(display_style="html");
```

```python
preus_2019 = np.array([ # array amb dimensions (bitllet, bonificació, zona)
    # bitllet senzill
    [[2.20, 3.10, 4.10, 5.25, 6.70, 7.80] for _ in range(3)], # totes 3 bonificacions, de zona 1 a 6
    # T-10
    [[10.20, 20.20, 27.40, 35.25, 40.50, 43.05] for _ in range(3)], # totes 3 bonificacions, de zona 1 a 6
    [ # T-mes
        [54, 72.70, 102, 124.90, 143.35, 153.55], # sense
        [43.20, 58.15, 81.60, 99.95, 114.65, 122.85], # FM/FM general
        [27.00, 36.35, 51.00, 62.45, 71.70, 76.80] # FM/FN especial
    ],
    [ # T-Trimestre
        [145.30, 196.50, 275.25, 337.15, 386.80, 414.40],
        [116.25, 157.20, 220.20, 269.75, 309.45, 331.55],
        [72.65, 98.25, 137.65, 168.60, 193.40, 207.20]
    ],
    [ # T-jove
        [105.00, 142.00, 199.20, 244.00, 280.00, 300.00],
        [84.00, 113.60, 159.35, 195.20, 224.00, 240.00],
        [52.50, 71.00, 99.60, 122.00, 140.00, 150.00]
    ],
    [ # T-70/30
        [60.90, 88.05, 120.75, 147.90, 169.35, 183.70],
        [57.05, 112.35, 153.30, 197.40, 226.65, 240.95],
        [35.65, 70.20, 95.80, 123.40, 141.65, 150.60]
    ]
])
coords_2019 = {
    "bitllet": [
        "senzill", "t-10", "t-mes", "t-trimestre", "t-jove", "t-70/30"
    ],
    "bonificacio": ["cap", "fm/fn general", "fm/fn especial"],
    "zona": np.arange(1,7)
}
preus_2019 = xr.DataArray(
    preus_2019, 
    dims=["bitllet", "bonificacio", "zona"], 
    coords=coords_2019
)
```

```python
preus_2019.to_netcdf("preus_atm.nc", group="2019", mode="w")
```
